# Interpretación con Intervalos de Confianza

Autocorrelación (ACF):
         Los valores de autocorrelación que caen fuera de los intervalos de confianza (bandas azules) son considerados estadísticamente significativos. En este gráfico, los primeros 3 a 4 retardos tienen valores significativos, lo que indica que hay una correlación fuerte entre estos retardos y los valores actuales.
 Decay lento: La decadencia lenta en la autocorrelación sugiere la presencia de una tendencia en los datos.

Autocorrelación Parcial (PACF):
        Significancia: Los primeros 2 a 3 retardos tienen valores significativos de autocorrelación parcial, ya que están fuera de los intervalos de confianza. Esto indica que estos retardos son importantes para explicar la serie temporal sin la influencia de los retardos intermedios.
        Corte brusco: Después de los primeros 2 a 3 retardos, los valores de autocorrelación parcial caen rápidamente dentro de los intervalos de confianza, lo que sugiere que un modelo AR con 2 o 3 términos puede ser adecuado.

# Resumen de la Interpretación

Tendencia: La lenta decadencia en la ACF sugiere que hay una tendencia en los datos.
Modelo AR: La significancia de los primeros 2 a 3 retardos en el PACF sugiere que un modelo AR(2) o AR(3) puede ser adecuado para capturar la estructura temporal de la serie.
Autocorrelación significativa

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from skforecast.plot import plot_residuals
import plotly.graph_objects as go
import plotly.io as pio
import plotly.offline as poff
pio.templates.default = "seaborn"
poff.init_notebook_mode(connected=True)
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams.update({'font.size':8})
```

```python
datos=pd.read_csv(r"C:\Users\arzua\OneDrive\Escritorio\preelec\src\data\datos_readaptados.csv")
```

```python

fig, ax = plt.subplots(figsize=(5, 2))
plot_acf(datos.Demand, ax=ax, lags=60)
plt.show()

fig, ax = plt.subplots(figsize=(5, 2))
plot_pacf(datos.Demand, ax=ax, lags=60)
plt.show()

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```
